### OHE with only Top Categories(Most Frequent Categories)

```python
counts = df['brand'].value_counts()
counts
```

```python
threshold = 100
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```
