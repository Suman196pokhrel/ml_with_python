```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../datasets/train_housing.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
```

```python
df['GarageQual'].mode()
```

```python
fig, ax = plt.subplots()

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax,color='red',linewidth=7,alpha=0.4)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='orange',linewidth=7,alpha=0.4)


lines, labels = ax.get_legend_handles_labels()
labels = ["Houses with TA","Houses with NA"]
ax.legend(lines, labels,loc='best')
```
