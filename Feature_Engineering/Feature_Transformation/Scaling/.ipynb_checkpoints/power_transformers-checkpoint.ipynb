```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('../datasets/concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df.drop(columns=['Strength'])
Y = df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

### Applying Regression without any Transformation

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = lr.predict(X_test)
r2_score(Y_test,Y_pred)
```

#### COnfirming the score using Cross validation

```python
lr = LinearRegression()
np.mean(cross_val_score(lr,X,Y,scoring='r2'))
```

#### Plots without any transformations

```python
for col in X_train.columns:
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=(14,4))
    
    sns.histplot(X_train[col],ax=ax1,kde=True,alpha=0.3)
    ax1.set_title(col)
    
    stats.probplot(X_train[col], dist="norm",plot=ax2)
    ax2.set_title(col)
```

### Now Lets Apply Box-Cox

```python
pt = PowerTransformer(method='box-cox')

X_train_transformed = pt.fit_transform(X_train+0.000001)
X_test_transformed = pt.transform(X_test+0.000001)

pd.DataFrame({'cols':X_train.columns, 'box_cox_lambdas':pt.lambdas_})
```

### Applying LInear Regression on Transformed Data

```python
lr_transformed = LinearRegression()
lr_transformed.fit(X_train_transformed,Y_train)

y_pred_2 = lr_transformed.predict(X_test_transformed)
r2_score(Y_test,y_pred_2)
```

```python
### using Cross val
pt = PowerTransformer(method='box-cox')
X_transformed = pt.fit_transform(X+ 0.000001)

lr = LinearRegression()
np.mean(cross_val_score(lr,X_transformed,Y,scoring='r2'))
```

#### AS you can see, there is a significant increase in Accuracy

### Lets See the Plots After Transformation

```python
X_train_transformed = pd.DataFrame(X_train_transformed,columns=X_train.columns)

for col in X_train_transformed:
    fig, (ax1, ax2) = plt.subplots(1,2,figsize=(14,4))
    
    sns.histplot(X_train[col],ax=ax1,kde=True,alpha=0.3)
    ax1.set_title(col)
    
    sns.histplot(X_train_transformed[col],ax=ax2,kde=True,alpha=0.3)
    ax2.set_title(col)
```

#### Look at the Age Feature, completly right skewed Data is now normalized

#### Using Yeo-Jhonson Transformer

```python
# default method is yeo-Jhonson
pt = PowerTransformer()

X_train_transformed_2 = pt.fit_transform(X_train)
X_test_transformed_2 = pt.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_transformed_2,Y_train)


y_pred_3 = lr.predict(X_test_transformed_2)
print(r2_score(Y_test,y_pred_3))
```

#### Applying CrossVal

```python
pt = PowerTransformer()

X_transformed_2 = pt.fit_transform(X)

lr = LinearRegression()
np.mean(cross_val_score(lr,X_transformed_2,Y,scoring='r2'))
```

#### We started from 0.46 and now we are at 0.68 , now thats power of powertransformations
