## We can see that these two features are related in a linear fashion so , we can use simple linear regression to predict the value of y using x

# Y = MX + C

```python
def linear_regression(x,y):
    N = len(x)
    x_mean = x.mean()
    y_mean = y.mean()
    
    M_numerator = ((x - x_mean) * (y - y_mean)).sum()
    M_denominator = ((x - x_mean)**2).sum()
    
    M = M_numerator / M_denominator
    
    C = y_mean - (M * x_mean)
    
    regression_line = f'Y = {M} X + {C}'
    return regression_line , [M,C]
```

```python
linear_regression(x,y)
```

```python
m,c = linear_regression(x,y)[1]

plt.scatter(x,y,color='green')
plt.plot(x, m*x + c, color='orange',linestyle='dotted',linewidth=5)
```
