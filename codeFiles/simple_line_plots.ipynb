```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Simple linear plots

### Setting up out visualization themes

```python
plt.style.use("seaborn-whitegrid")
```

### Intro to FIgure and Axes instances

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0,10,1000)
```

```python
plt.plot(x, np.sin(x))
```

```python
ax.plot(x, np.sin(x))
fig
```

```python
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
```

```python
ax.plot(x, np.cos(x))
fig
```

## Adjusting the styles

```python
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color='0.75')
plt.plot(x,np.sin(x-3),color='#FFDD44')
plt.plot(x,np.sin(x-3),color=(1.0,0.2,0.3))  ## RGB tuple
```

### Line styles

```python
plt.plot(x,x-0,linestyle='solid')
plt.plot(x,x-1,linestyle='dashed')
plt.plot(x,x-2,linestyle='dotted')
plt.plot(x,x-3,linestyle='dashdot')
```

## Adjusting the Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.axis([-1,11,-1.5,1.5])
```

## Labeling Plots

```python
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

## Plot Legends for multiple lines

```python
plt.plot(x, np.sin(x), color="green",linestyle = "solid",label='Sin(x)')
plt.plot(x, np.cos(x), color="blue",linestyle = "dashed",label='Cos(x)')
plt.plot(x, np.sin(x+5), color="orange",linestyle = "dashdot",label='Sin(x+5)')



plt.axis('equal')
plt.legend()
```
