```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("seaborn-whitegrid")
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
```

## 1st method : Using plt.plot() , but with a manual marker

```python
plt.plot(x,y,'o',color="black")
```

```python
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(np.random.randint(10), np.random.randint(10),marker)
```

```python
# Using Line, Circle and Colors in one single keyword

plt.plot(x, y,'-ok')
```

## 2nd Method: Using plt.Scatter()

```python
x,y
```

```python
plt.scatter(x,y,marker='o',color="orange")
plt.scatter(x,np.cos(x),marker='<',color='blue')
```
