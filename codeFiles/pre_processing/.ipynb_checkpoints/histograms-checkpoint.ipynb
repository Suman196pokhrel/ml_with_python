```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-white')
```

# One - Dimensional Histogram

```python
#  Here were using randn() function to generate a normally distributed 1000 numbers
data = np.random.randn(2000)
```

```python
plt.hist(data)
```

## Styling histogram

```python
plt.hist(data,bins=20,alpha=0.7,color='orange',histtype='stepfilled',edgecolor='black')
```

# Example 3

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, bins=40)
```

```python
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

# Two - Dimensional Histogram

## Example 1

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
```
