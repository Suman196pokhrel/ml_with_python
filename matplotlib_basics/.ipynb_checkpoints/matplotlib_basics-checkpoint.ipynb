# Simple Line PLots

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,20,100)

# Customizing the plot lines
ax.plot(x,np.sin(x),'*',color='green')
ax.plot(x,np.cos(x-1),'--',color='red')
ax.plot(x,np.cos(x-2),'.',color='yellow')

# Changing the plot limits
ax.set_xlim(-5,25)
# ax.set_ylim(0,1.5)

# Labeling Plots
ax.set_title("SIn & Cosine Plots")
ax.set_xlabel("deciams from -5 to 25")
ax.set_ylabel("Tignometric values from -1 to 1")
```
