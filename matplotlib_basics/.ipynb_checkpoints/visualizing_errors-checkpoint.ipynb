```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Basic ErrorBars

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

fig, ax = plt.subplots()

ax.errorbar(x,y,yerr=dy,fmt=".k",capsize=2.5)
```

```python
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr=dy, fmt='o', color='black',ecolor='lightgray', elinewidth=3, capsize=0)
```
