```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## SImple scatter plots

```python
x = np.linspace(0,10,100)
y = np.sin(x)

fig, ax1 = plt.subplots()

ax1.plot(x,y,"*")
```

```python
fig, ax1 = plt.subplots()

rng = np.random.RandomState(42)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    ax1.plot(rng.rand(5),rng.rand(5),marker,label=f"marker = {marker}")

ax1.legend(loc="upper right")
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x,y,'-p',color='gray',markersize=10, linewidth=1,markerfacecolor="white",markeredgecolor="gray",markeredgewidth=2)
```

# Using ax1.scatter() => gives control over each indivisual points

```python
fig,ax1 = plt.subplots()
ax1.scatter(x,y,marker='o')
```

```python
fig,ax1 = plt.subplots()

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors_1 = rng.rand(100)
sizes = 1000 * rng.rand(100)

ax1.scatter(x,y,c=colors_1,s=sizes,alpha=0.3)

# Creating color bars
fig.colorbar(plt.cm.ScalarMappable(cmap="viridis"),ax=ax1)
```
