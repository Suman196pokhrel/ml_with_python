### Assuming fuel to be Nominal categorical variable we will be using one hot encoding instead of label encoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def handle_cat(df,feature_list):
    for feature_name in feature_list:
        ohe = OneHotEncoder()
        df_ohe = ohe.fit_transform(df[[feature_name]])
        df_ohe = df_ohe.toarray() # convertinf sparse matrix to array
        ohe_df = pd.DataFrame(df_ohe,columns=ohe.categories_[0]).iloc[:,1:]   # converting array to DF and dropping 1st column to avoid dummy_var_trap
        df = pd.concat([df.drop(feature_name,axis=1), ohe_df.astype(int)],axis=1)   # concatenating to main df
    
    return df
```

```python
feature_list = df.select_dtypes(include=[np.object0]).columns.values
feature_list = feature_list.tolist()
feature_list
```

```python
# Handleing all Cat values
ohe_df = handle_cat(df=df,feature_list=feature_list)
ohe_df.head()
```

```python
ohe_df.shape
```

### Lets do the same using sklearn

```python
from sklearn.compose import ColumnTransformer

cat_pipeline = ColumnTransformer([
    ("cat",OneHotEncoder(),feature_list)
])

ohe_using_pipelines = cat_pipeline.fit_transform(df)
ohe_using_pipelines.toarray().shape
```
