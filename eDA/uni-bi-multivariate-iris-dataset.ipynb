# Univariate , Bivariate and Multivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/IRIS.csv")
df.sample(5)
```

```python
df.shape
```

```python
df["species"].value_counts()
```

## Univariate Analysis

```python
df_setosa = df.loc[df['species']=='Iris-setosa']
```

```python
df_virginica = df.loc[df['species']=='Iris-virginica']
df_versicolor = df.loc[df['species']=='Iris-versicolor']
```

```python
plt.plot(df_setosa["sepal_length"],np.zeros_like(df_setosa["sepal_length"]),'o')
plt.plot(df_versicolor["sepal_length"],np.zeros_like(df_versicolor["sepal_length"]),'o')
plt.plot(df_virginica["sepal_length"],np.zeros_like(df_virginica["sepal_length"]),'o')
plt.xlabel("Sepal Length")
```

## Bivariate Analysis

```python
fig,ax = plt.subplots()

ax.scatter(x=df_setosa["sepal_length"],y=df_setosa["sepal_width"],label="Setosa")
ax.scatter(x=df_versicolor["sepal_length"],y=df_versicolor["sepal_width"],label="VersiColor")
ax.scatter(x=df_virginica["sepal_length"],y=df_virginica["sepal_width"],label="Virginica")
fig.legend(loc='upper right', bbox_to_anchor=(0.4, 0.37, 0.5, 0.5))

ax.set_xlabel("Sepal_Length")
ax.set_ylabel("Sepal_Width")
```
